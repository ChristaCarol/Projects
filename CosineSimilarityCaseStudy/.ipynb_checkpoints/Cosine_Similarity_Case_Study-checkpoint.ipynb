## Cosine Similarity Calculations
Cosine similarity is a measure of similarity between two non-zero vectors of an inner product space that measures the cosine of the angle between them. Similarity measures have a multiude of uses in machine learning projects; they come in handy when matching strings, measuring distance, and extracting features. This similarity measurement is particularly concerned with orientation, rather than magnitude. 
In this case study, you'll use the cosine similarity to compare both a numeric data within a plane and a text dataset for string matching.

Load the Python modules, including cosine_similarity, from sklearn.metrics.pairwise

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.cm as cm
%matplotlib inline
plt.style.use('ggplot')
from scipy import spatial
from sklearn.metrics.pairwise import cosine_similarity
```

### Cosine Similarity with clusters and numeric matrices

All points in our dataset can be thought of as feature vectors. We illustrate it here as we display the __Cosine Similarity__ between each feature vector in the YZ plane and the [5, 5] vector we chose as reference. The sklearn.metrics.pairwise module provides an efficient way to compute the __cosine_similarity__ for large arrays from which we can compute the similarity.

 **<font color='teal'> First, create a 2D and a 3D matrix from the dataframe. The 2D matrix should contain the 'Y' and 'Z' columns and the 3D matrix should contain the 'X','Y', and 'Z' columns.</font>**

```python
# Create a 2D matrix with 'Y' and 'Z' columns
YZ_matrix = data[['Y', 'Z']].values

# Create a 3D matrix with 'X', 'Y', and 'Z' columns
XYZ_matrix = data[['X', 'Y', 'Z']].values

# Print the first few rows of each matrix to verify
print("First few rows of the 2D YZ matrix:")
print(YZ_matrix[:5])
print("\nFirst few rows of the 3D XYZ matrix:")
print(XYZ_matrix[:5])
```

Calculate the cosine similarity for those matrices with reference planes of 5,5 and 5,5,5. Then subtract those measures from 1 in new features.

```python
# Calculate the cosine similarity for the 3D matrix with reference [5, 5, 5]
simCosine3D = 1. - cosine_similarity(XYZ_matrix, [[5, 5, 5]])

# Calculate the cosine similarity for the 2D matrix with reference [5, 5]
simCosine2D = 1. - cosine_similarity(YZ_matrix, [[5, 5]])

# Print the results to check
print("Cosine Similarity from 1 for 3D matrix:")
print(simCosine3D[:5])  # Print first 5 results to check

print("\nCosine Similarity from 1 for 2D matrix:")
print(simCosine2D[:5])  # Print first 5 results to check
```

Using the 2D matrix and the reference plane of (5,5) we can use a scatter plot to view the way the similarity is calculated using the Cosine angle.

```python
# Create a scatter plot of the YZ matrix
plt.figure(figsize=(8, 8))
plt.scatter(YZ_matrix[:, 0], YZ_matrix[:, 1], alpha=0.5, label='Data Points')

# Plot the reference vector [5, 5]
plt.scatter([5], [5], color='red', s=100, label='Reference Vector (5,5)')

# Optionally, draw lines from a few points to the reference vector to illustrate the angles
# Here, we're drawing lines from the first 10 points to the reference vector
for point in YZ_matrix[:10]:
    plt.plot([point[0], 5], [point[1], 5], 'gray', linestyle='--', alpha=0.5)

plt.title('Cosine Similarity Visualization in 2D')
plt.xlabel('Y Coordinate')
plt.ylabel('Z Coordinate')
plt.legend()
plt.grid(True)
plt.show()
```

Now, plot the 3D matrix with the similarity and the reference plane, (5,5,5).

```python
from mpl_toolkits.mplot3d import Axes3D

# Set up a 3D plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Scatter plot for the XYZ matrix
ax.scatter(XYZ_matrix[:, 0], XYZ_matrix[:, 1], XYZ_matrix[:, 2], alpha=0.5, label='Data Points')

# Plot the reference vector [5, 5, 5]
ax.scatter([5], [5], [5], color='red', s=100, label='Reference Vector (5,5,5)')

# Optionally, draw lines from a few points to the reference vector to illustrate the angles
# Here, we're drawing lines from the first 5 points to the reference vector
for point in XYZ_matrix[:5]:
    ax.plot([point[0], 5], [point[1], 5], [point[2], 5], 'gray', linestyle='--', alpha=0.5)

ax.set_title('Cosine Similarity Visualization in 3D')
ax.set_xlabel('X Coordinate')
ax.set_ylabel('Y Coordinate')
ax.set_zlabel('Z Coordinate')
ax.legend()
plt.show()
```

### Cosine Similarity with text data
This is a quick example of how you can use Cosine Similarity to compare different text values or names for record matching or other natural language proecessing needs. 
First, we use count vectorizer to create a vector for each unique word in our Document 0 and Document 1. 

```python
from sklearn.feature_extraction.text import CountVectorizer
count_vect = CountVectorizer()
Document1 = "Starbucks Coffee"
Document2 = "Essence of Coffee"

corpus = [Document1,Document2]

X_train_counts = count_vect.fit_transform(corpus)

df = pd.DataFrame(X_train_counts.toarray(), columns=count_vect.get_feature_names_out(), index=['Document 0', 'Document 1'])
```

Now, we use a common frequency tool called TF-IDF to convert the vectors to unique measures.

```python
from sklearn.feature_extraction.text import TfidfVectorizer
# Define the documents
Document1 = "Starbucks Coffee"
Document2 = "Essence of Coffee"

# Create a corpus with the documents
corpus = [Document1, Document2]

# Initialize the TfidfVectorizer
vectorizer = TfidfVectorizer()

# Fit and transform the corpus to TF-IDF features
trsfm = vectorizer.fit_transform(corpus)

# Create a DataFrame from the vectors to display them more clearly
df = pd.DataFrame(trsfm.toarray(), columns=vectorizer.get_feature_names_out(), index=['Document 0', 'Document 1'])

# Display the DataFrame
print(df)
```

Here, we finally apply the __Cosine Similarity__ measure to calculate how similar Document 0 is compared to any other document in the corpus. Therefore, the first value of 1 is showing that the Document 0 is 100% similar to Document 0 and 0.26055576 is the similarity measure between Document 0 and Document 1.

```python
cosine_similarity(trsfm[0:1], trsfm)
```

Replace the current values for `Document 0` and `Document 1` with your own sentence or paragraph and apply the same steps as we did in the above example.

 **<font color='teal'> Combine the documents into a corpus.</font>**

```python
Document0 = "Today is a beautiful day to go fishing and enjoy the glorious sunshine."
Document1 = "I don't know about you, but I am excited for summer to finally be here."

# Create a corpus with the documents
corpus = [Document0, Document1]

# Initialize the TfidfVectorizer
vectorizer = TfidfVectorizer()

# Fit and transform the corpus to TF-IDF features
trsfm = vectorizer.fit_transform(corpus)

# Create a DataFrame from the vectors to display them more clearly
df = pd.DataFrame(trsfm.toarray(), columns=vectorizer.get_feature_names_out(), index=['Document 0', 'Document 1'])

# Compute cosine similarity
similarity_scores = cosine_similarity(trsfm[0:1], trsfm)

# Display the DataFrame showing TF-IDF vectors
print("TF-IDF Matrix:")
print(df)
print("\nCosine Similarity Scores:")
print(similarity_scores)
```
